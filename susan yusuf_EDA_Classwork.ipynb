```python
nums = np.array([1, 3, 5, 7, 9])
print("array:", nums)
```

```python
nums = np.sum(nums)
print(nums)
```

```python
nums= np.max(nums)
print(nums)
```

```python
nums = np.mean(nums)
print(nums)
```
